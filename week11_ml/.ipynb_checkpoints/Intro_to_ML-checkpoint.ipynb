```python
import os
import numpy as np
import pandas as pd

os.chdir("/Users/brandonye/Desktop/fa22/QBIO 490/fa22_qbio_490_brandon/analysis_data")
```

```python
import cptac

cptac.download(dataset="Brca")
brca = cptac.Brca()
protein_data = brca.get_proteomics()
protein_data = protein_data.dropna(axis=1)
```

```python
# Elbow method to find ideal number of clusters
# Plot clusters vs. Within Cluster Sum of Squares (WCSS)
# implementation from https://towardsdatascience.com/machine-learning-algorithms-part-9-k-means-example-in-python-f2ad05ed5203

from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

wcss = []
for i in range(1, 11): # num clusters
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(protein_data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show() # elbow point at 4
```

```python
# PCA using TCGA RNA data
rna = pd.read_csv("/Users/brandonye/Desktop/fa22/QBIO 490/fa22_qbio_490_brandon/analysis_data/brca_rna_count_data.csv", index_col = 0)
clinical = pd.read_csv("/Users/brandonye/Desktop/fa22/QBIO 490/fa22_qbio_490_brandon/analysis_data/brca_rna_clinical_data.csv", index_col = 0)
```

```python
rna.head()
clinical.head()
```

```python
# implementation from https://towardsdatascience.com/pca-using-python-scikit-learn-e653f8989e60

from sklearn.preprocessing import StandardScaler

rna = rna.transpose()
features = rna.columns

# Separating out the features
x = rna.loc[:, features].values
y = clinical.loc[:, ["paper_BRCA_Subtype_PAM50"]].values

# Standardizing the features
scaled_rna = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(scaled_rna)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])

principalDf["paper_BRCA_Subtype_PAM50"] = y
```

```python
clinical["paper_BRCA_Subtype_PAM50"].unique()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ["Basal", "LumA", "Normal", "Her2", "LumB", "nan"]
colors = ['r', 'b', 'g', 'c', 'm', 'y']
for target, color in zip(targets,colors):
    indicesToKeep = principalDf["paper_BRCA_Subtype_PAM50"] == target
    ax.scatter(principalDf.loc[indicesToKeep, 'principal component 1']
               , principalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
# K-means clustering
# implementation from https://towardsdatascience.com/machine-learning-algorithms-part-9-k-means-example-in-python-f2ad05ed5203

kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(protein_data)
plt.scatter(protein_data.iloc[:,0], protein_data.iloc[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.title("K-means Clustering on CPTAC BRCA Proteomic Data (n_clusters = 4)")
plt.show()
```

```python
# UMAP for RNA data colored by K-means Clusters
# implementation from https://umap-learn.readthedocs.io/en/latest/basic_usage.html

from sklearn.preprocessing import OneHotEncoder
import sklearn.cluster as cluster
import umap

encoder = OneHotEncoder(handle_unknown = "ignore") # encode categorical variables
clinical = encoder.fit_transform(clinical).toarray()

kmeans_labels = cluster.KMeans(n_clusters=4).fit_predict(clinical)
k_means_embedding = umap.UMAP().fit_transform(clinical)

plt.scatter(k_means_embedding[:, 0], k_means_embedding[:, 1], c=kmeans_labels, s=1, cmap='Spectral');
plt.title("UMAP for RNA Data (Colored by K-means Clusters)")
plt.show()
```

```python
# UMAP for RNA data colored by TP53 expression

encoder = OneHotEncoder(handle_unknown = "ignore") # encode categorical variables
tp_53_expression = np.where(protein_data.loc[:, 'TP53'] >= 1, 0, np.where(protein_data.loc[:, 'TP53'] <= -1, 1, 2))
protein_data = encoder.fit_transform(protein_data).toarray()

tp53_embedding = umap.UMAP().fit_transform(protein_data)

plt.scatter(tp53_embedding[:, 0], tp53_embedding[:, 1], c=tp_53_expression, s=1, cmap='Spectral');
plt.title("UMAP for RNA Data (Colored by TP53 Expression)")
plt.show()
```
